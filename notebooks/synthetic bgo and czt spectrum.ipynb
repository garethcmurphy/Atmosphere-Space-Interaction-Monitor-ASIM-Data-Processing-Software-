```python
%matplotlib inline 
import numpy as np
import pylab
import matplotlib.pyplot as plt
#plt.tight_layout()

ymax=400
ymin=1e-1
xmax=40
xmin=0.1
fpar = np.linspace(0.1,0.4,5)
#pylab.axis([2,20,0,14])

from pylab import *
import matplotlib.pyplot  as pyplot
a = [ pow(10,i) for i in range(10) ]
fig = pyplot.figure(figsize=(12, 12))
fig.subplots_adjust(bottom=0.1)
fig.suptitle("MXGS CZT and BGO spectra", fontsize=14)
for i in range(4):
    x = fpar
    y = ymax* np.exp(-fpar/3.)
    y = y*(1 + (np.random.random_sample(y.shape)-0.5)*0.5)

    #A = np.vstack([x, np.ones(len(x))]).T
    #m, c = np.linalg.lstsq(A, y)[0]

    ax = fig.add_subplot(4,4,i+1)

    line, = ax.plot(x,y, 'r+')
    ax.set_autoscalex_on(False)
    ax.set_xlim([xmin,xmax])
    ax.set_autoscaley_on(False)
    ax.set_ylim([ymin,ymax*1.1])
    ax.set_xscale('log')
    ax.set_yscale('log')

    ax.set_title(" CZT"+str(i) )

fpar =  np.linspace(0.4,40,20)
for i in range(12):
    x = fpar
    y = ymax* np.exp(-fpar/3.)
    y = y*(1 + (np.random.random_sample(y.shape)-0.5)*0.5)
    
    #A = np.vstack([x, np.ones(len(x))]).T
    #m, c = np.linalg.lstsq(A, y)[0]

    ax = fig.add_subplot(4,4,4+i+1)

    line, = ax.plot(x,y, 'b+')
    ax.set_autoscalex_on(False)
    ax.set_xlim([xmin,xmax])
    ax.set_autoscaley_on(False)
    ax.set_ylim([ymin,ymax*1.1])
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_title("BGO "+str(i))
   
savefig('cztbgo.png')

show()
```
