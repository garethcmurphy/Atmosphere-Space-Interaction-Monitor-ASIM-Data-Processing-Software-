```python
#%matplotlib inline
from sys import argv
def is_interactive():
    import __main__ as main
    return not hasattr(main, '__file__')

if is_interactive():
    params = [1]
else:
    params = argv[1:]
    


def plot_fits_image(filename): 
    from astropy.io import fits
    hdulist = fits.open(filename)
    hdulist.info()
    #print hdulist[0].header
    scidata = hdulist[0].data
    import re
    regx = re.compile('fits\\b')
    outname=regx.sub('png',filename)
    import matplotlib.pyplot as plt
    import matplotlib as mpl
    from matplotlib import cm
    mpl.rcParams.update({'font.size': 5})
    mpl.rc('font',family='Times New Roman')
    fig=plt.figure(figsize=(3, 3.5))
    ax1=fig.add_subplot(111, aspect='equal')
    im=plt.imshow (scidata, extent=[-42,42,-42,42], cmap=cm.plasma)
    fig.suptitle('MXGS nadir FOV image for event: TGF_test_example')
    labelstr='Nadir %s-offset angle (deg)'
    ax1.set_ylabel(labelstr % 'y')
    ax1.set_xlabel(labelstr % 'x')
    cbar = plt.colorbar(im,fraction=0.046, pad=0.14, orientation='horizontal')
    plt.savefig(outname,  dpi=300)
    plt.close()





```

```python

def plot_mdap_dat(fname):
    from astropy.io import ascii
    table =fname
    data = ascii.read(table)
    print data.meta['comments']
    comments=data.meta['comments']
    headings=comments[-2]
    heads=headings.split()
    unitsstr=comments[-1]
    units=unitsstr.split()
    print units
    import re
    regx = re.compile('dat\\b')
    outname=regx.sub('png',fname)
    time=data['col1']
    count=data['col2']
    import matplotlib.pyplot as plt
    import matplotlib as mpl
    from matplotlib import cm
    mpl.rcParams.update({'font.size': 10})
    mpl.rc('font',family='Times New Roman')
    fig=plt.figure(figsize=(4, 3))
    ax1=fig.add_subplot(111)
    plt.plot (time, count)
    plottitle=comments[1]
    ytitle=heads[1]+' '+units[1]
    xtitle=heads[0]+' '+units[0]
    fig.suptitle(plottitle)
    ax1.set_ylabel(ytitle)
    ax1.set_xlabel(xtitle)
    fig.tight_layout()
    plt.savefig(outname,  dpi=300)
    plt.close()



```

```python
def get_dm_locs():
    dmsize=4.6
    off=0.5
    dau_lat=0.9
    dm_lat=0.25
    import numpy as np
    arr=np.array([0+dau_lat/2,dmsize+dau_lat/2+dm_lat ,2*dmsize+dau_lat/2+2*dm_lat ,3*dmsize+dau_lat/2+3*dm_lat ])
    b=np.concatenate((-arr[::-1]-dmsize,arr), axis=0)
    return b
```

```python

def scatter_plot_detector(fname):
    from astropy.io import ascii
    table =fname
    data = ascii.read(table)
    print data.meta['comments']
    import re
    regx = re.compile('dat\\b')
    outname=regx.sub('png',fname)
    time=data['col5']
    count=data['col6']
    import matplotlib.pyplot as plt
    import matplotlib.patches as patches
    import matplotlib as mpl
    from matplotlib import cm
    mpl.rcParams.update({'font.size': 9})
    mpl.rc('font',family='Times New Roman')
    fig=plt.figure(figsize=(3, 3.5))
    ax1=fig.add_subplot(111, aspect='equal')
    ax1=fig.add_subplot(111, aspect='equal')
    locs=get_dm_locs()
    for i in locs:
        for j in locs:
            ax1.add_patch(
            patches.Rectangle(
                (i, j),   # (x,y)
                4.6,          # width
                4.6,          # height
                facecolor="chartreuse",
                zorder=1
            )
                 )
    plt.scatter (time, count, zorder=3)
    #plt.axes().set_aspect('equal', 'datalim')
    plt.ylim(-20,20)
    plt.xlim(-20,20)
    fig.suptitle('MXGS detector count pattern')
    axlabel='Detector %s-offset (cm)'
    ax1.set_ylabel(axlabel % 'Y' )
    ax1.set_xlabel(axlabel % 'X')
    fig.tight_layout()
    plt.savefig(outname,  dpi=300)
    plt.close()



path='/Users/garmu/Dropbox/asdc/level2/mxgs/MDAP/TGF_test_example/Output/'

path='./'

    
fname=path+'MDAP_image_nadir_fov.fits'
plot_fits_image(fname)


    
fname=path+'MDAP_image_correlation.fits'
plot_fits_image(fname)

# path='/Users/garmu/Documents/TGF_test_example/Output/'


# fname=path+'MDAP_image_background.fits'

# plot_fits_image(fname)
    
    
fname=path+'MDAP_detector_event_list.dat'
scatter_plot_detector(fname)




fname=path+'MDAP_detector_count_lightcurve.dat'
plot_mdap_dat(fname)


fname=path+'MDAP_detector_count_spectrum.dat'
plot_mdap_dat(fname)
```
