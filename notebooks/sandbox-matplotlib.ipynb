```python
%matplotlib inline

# plotfile.py
from mpl_toolkits.basemap import Basemap

import matplotlib.pyplot as plt


class Plotter(object):
    def __init__(self, xval=None, yval=None):
        self.xval = xval
        self.yval = yval

    def plotthing(self):
        f = plt.figure()
        lonarr=[-60,1,60]
        latarr=[25,30,70]
        minlat= -60
        maxlat= 60
        minlon= -179
        maxlon= 179
        lightningmap =  Basemap(projection='merc',llcrnrlat=minlat,urcrnrlat=maxlat,\
                llcrnrlon=minlon,urcrnrlon=maxlon,lat_ts=20,resolution='c')
        x, y = lightningmap(lonarr, latarr) 
        lightningmap.drawcoastlines()
        lightningmap.drawcountries()
        lightningmap.plot(x, y, 'D-', markersize=4, linewidth=2, color='k', markerfacecolor='b') 
        return f





#from plotfile import Plotter

app = Plotter(xval=range(0,10), yval=range(0,10))
plot = app.plotthing()
#plot.show()
```

```python
%matplotlib inline
import matplotlib.pyplot  as plt
class dataClass(object):
    def __init__(self, name, dictionary):
        self.name = name
        self.models = {}
        self.add_model(dictionary)
        self.ax1=range(1,10)
        self.ax2=range(1,10)
        

    def add_model(self, dictionary):
        model_name = dictionary['model']
        # get django stuff here
        if 'data' in dictionary:
            data = dictionary['data']
        else:
            data = None
        self.models.update({model_name : data})

class model(object):
    def __init__(self, model_name):
        self.modelname = model_name

    def plot(self, thing_to_plot , fig=None, ax=111, xaxis=None, **kwargs):
        # do meaningful work here
        plt.plot (thing_to_plot.ax1)
        return thing_to_plot.name

example = dataClass('Aluminum', {'model': 'Thermal Conductivity'})
thing = model("doug")

print("Plotted object's name: %s" % thing.plot(example))
print("_model name: %s" % thing.modelname)
```
